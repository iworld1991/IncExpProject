## An example for consumption policy (a function of s and ϵ)

```python
s_grid_size = 100
ϵ_grid_size = 100
n = 10 
```

```python
@njit
def mulfunc(s,ϵ):
    val = np.sin(2*s**0.3*ϵ**(0.7))
    return val 
```

```python
a_grid = np.linspace(0.0,2.0,s_grid_size)
ϵ_grid = np.linspace(0.0,2.0,ϵ_grid_size)
```

```python
aa,ϵϵ = np.meshgrid(a_grid,
                    ϵ_grid)
```

```python
#σ_in = np.array([[mulfunc(u1,u2) for u2 in x2] for u1 in x1])
σ_in = mulfunc(aa,
               ϵϵ)
```

```python
σ_in.shape
```

```python
plt.plot(σ_in[1,:])
plt.plot(σ_in[8,:])
```

```python
grid = (a_grid,ϵ_grid)
```

```python
σ_func =lambda s,ϵ: mlinterp(grid, σ_in.T, (s, ϵ))
```

```python
val_fun =[]
val_int = []

s_ls = np.linspace(0,2,10)

for s in s_ls:
    val_int.append(σ_func(s,0.1))
    val_fun.append(mulfunc(s,0.1))
    
    
    
plt.plot(val_fun,s_ls,label='int')
plt.plot(val_int,s_ls,label='func')
plt.legend(loc=0)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_surface(aa, 
                ϵϵ,
                σ_in, 
                rstride=1, 
                cstride=1,
                cmap='viridis', edgecolor='none'
               )
#ax.scatter(s_points, ϵ_points, σ_in_int)
#ax.set_xlabel('x')
#ax.set_ylabel('y')
#ax.set_zlabel('z');
```
